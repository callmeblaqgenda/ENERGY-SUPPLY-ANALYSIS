```python
# First thing to do is to load up the libaries 

import pandas as pd 
import numpy as np
```

### For Energy Indicators data:

```python
# Load up and clean the Energy Indicators data 

Energy = pd.read_excel( "Energy Indicators.xls", header=16, skipfooter=38, skiprows=1, usecols='C:F', na_values='...')
Energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

# Convert 'Energy Supply' to gigajoules (Note: 1,000,000 gigajoules = 1 petajoule)

Energy['Energy Supply'] *= 1_000_000  # <--- since the energy supply column is in petajoules then multiply it with 1_000_000

# Clean the country names to remove unnecessary numbers and parentheses 

Energy['Country'] = Energy['Country'].str.replace(r'\d+', '', regex=True) 
Energy['Country'] = Energy['Country'].str.replace(r'\(.*\)', '', regex=True)

# Rename some specific countries 

Energy['Country'] = Energy['Country'].replace({
    "Republic of Korea": "South Korea",
    "United States of America": "United States", 
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom", 
    "China, Hong Kong Special Administrative Region": "Hong Kong"
})

# Display the first 15 rows 

Energy.head(15) 
```

### For GDP data:

```python
# Load up and clean the GDP data 

GDP = pd.read_csv('world_bank.csv', header=1, skiprows=3) 

# Rename some specific countries 

GDP['Country Name'] = GDP['Country Name'].replace({
    "Korea, Rep.": "South Korea", 
    "Iran, Islamic Rep.": "Iran", 
    "Hong Kong SAR, China": "Hong Kong"
})

# Filter the GDP data from the last 10 years (2006-2015) 

GDP = GDP[['Country Name'] + [str(year) for year in range(2006, 2016)]]

# Display the first 15 rows 

GDP.head(15)
```

### For Scimagojr data:

```python
# Load up and clean the Scimagojr data 

ScimEn = pd.read_excel('scimagojr-3.xlsx') 

# Filter the Scimagojr data to the top 15 ranked countries 

ScimEn_top15 = ScimEn[ScimEn['Rank'] <= 15]

# Display the first 15 rows 

ScimEn.head(15)
```

### Merging the datasets:

```python
# To ensure the consistent formatting of the country names across the 3 datasets 

Energy['Country'] = Energy['Country'].str.strip().str.title()
GDP['Country Name'] = GDP['Country Name'].str.strip().str.title()
ScimEn['Country'] = ScimEn['Country'].str.strip().str.title() 

# Merging the datasets together based on 'Country'

first_merged = pd.merge(ScimEn_top15, Energy, how='inner', left_on='Country', right_on='Country')
final_merged = pd.merge(first_merged, GDP, how='inner', left_on='Country', right_on='Country Name')

# Setting the index of the dataframe and reordering the columns 

final_merged.set_index('Country', inplace=True) 
new_columns_order = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 
                     'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', 
                     '% Renewable'] + [str(year) for year in range(2006, 2016)] 

# Assigning the merged datasets into a new dataframe and finally sorting it by 'Rank' 

final_result = final_merged[new_columns_order]
final_result = final_result.sort_values(by='Rank') 

# Displaying the first 15 rows of my well-organized final result

final_result.head(15) 
```
